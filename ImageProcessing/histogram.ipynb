```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_image( infilename ) :
    img = Image.open( infilename )
    img.load()
    data = np.asarray( img, dtype="int32" )
    return data
```

```python
image_arr = load_image('./image.png')
```

```python
image_arr.shape
```

Fourth layer is for transparency, the first three are Red, Green and Blue

```python
image_arr = image_arr[...,:3]
```

```python
plt.imshow(image_arr)
```

```python
 image_arr
```

```python
red = image_arr[...,0]
green = image_arr[...,1]
blue = image_arr[...,2]
```

Vertical addition of values: axis=0  
Horizontal addition of values: axis=1

```python
red_vert = red.sum(axis=0)
green_vert = green.sum(axis=0)
blue_vert = blue.sum(axis=0)
```

```python
plt.plot(red_vert)
```

```python
plt.plot(green_vert)
```

```python
plt.plot(blue_vert)
```

```python
red_hort = red.sum(axis=1)
green_hort = green.sum(axis=1)
blue_hort = blue.sum(axis=1)
```

```python
plt.plot(red_hort)
```

```python
plt.plot(green_hort)
```

```python
plt.plot(blue_hort)
```
